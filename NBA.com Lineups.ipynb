```python
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline
import seaborn as sns
from collections import defaultdict

from sklearn import preprocessing
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

from sklearn.mixture import GaussianMixture as GMM
from sklearn.cluster import KMeans, SpectralClustering as Spectral
from sklearn.metrics import silhouette_samples, silhouette_score

import renders as rs
import reduce_data as rd

import tqdm
```

```python
# Import data
df = pd.DataFrame.from_csv('PositionClusters.csv')
df2 = pd.DataFrame.from_csv('Lineups.csv').reset_index(drop=True)
```

```python
df2 = df2[['Season', 'GROUP_ID', 'TEAM_ID', 'TEAM_ABBREVIATION', 'MIN', 'DEF_RATING', 'OFF_RATING', 'NET_RATING']]
df2.head()
```

```python
# Fix Season
seasons = []
for i in df2['Season']:
    i = int(i.split('-')[0]) + 1
    seasons.append(i)
df2['Season'] = seasons
df2.head()
```

```python
# Format IDs
PLAYER_IDS =[]
for i in df2['GROUP_ID']:
    i = i.replace(' -', ', ')
    PLAYER_IDS.append([int(s) for s in i.split(',')])
df2['PLAYER_IDS'] = PLAYER_IDS
df2.head()
```

```python
# Takes ~2 minutes: tqdm progress bar included
# Match IDs to Lineups
matches = []
for i in tqdm.tqdm(range(len(df))):
    for j in range(len(df2)):
        if (df['PLAYER_ID'][i] in df2['PLAYER_IDS'][j]) & (df['Season'][i] == (df2['Season'][j])):
            matches.append((j, i))
print matches            
```

```python
# Replace player numbers with their names and offensive positions
name_lineups = []
position_lineups = []
for i in matches:
    lineup = i[0]
    player = i[1] 
    name = df['Name'][player]
    position = df['Offensive Position'][player]
    name_lineups.append((lineup, name))
    position_lineups.append((lineup, position))
print position_lineups
```

```python
# Create dictionary to then count number of positions for each lineup

name_dict = defaultdict(list)
for k, v in name_lineups:
    name_dict[k].append(v)

position_dict = defaultdict(list)
for k, v in position_lineups:
    position_dict[k].append(v)

# Add player names
player_names = []
for i in name_dict:
    player_names.append(name_dict[i])
df2['PLAYER_NAMES'] = player_names

# If match, confirm its validity
valid_names = []
for i in df2['PLAYER_NAMES']:
    valid_names.append(len(i))
df2['VALID_NAMES'] = valid_names
df2.head()

# Add positions
positions = []
for i in position_dict:
    positions.append(position_dict[i])
df2['Positions'] = positions

# Count positions
position_list_counter = []
from collections import Counter
for i in position_dict:
    position_list_counter.append(Counter(position_dict[i]))
```

```python
# Create basic positions
basic_type = []
for i in position_list_counter:
    if i['Big'] >= 2:
        basic_type.append('Traditional')
    else:
        basic_type.append('Small-ball')
basic_type

df2['Basic_Type'] =  basic_type

# Valid only if 5 positions are matched
df = df2.loc[df2['VALID_NAMES'] == 5]
df.head()
```

```python
# NBA.com boxplot prefers small, but perhaps this should account for minutes
df.boxplot(column='NET_RATING', by='Basic_Type', figsize=(12, 8))
```

```python
# Adding weights and then grouping each lineup type by team-season produces adjusted ratings
g = df.groupby(['TEAM_ID', 'Season', 'Basic_Type'])

df['MINxNET'] = (df.MIN * df.NET_RATING) / (g.MIN.transform("sum"))
df['ADJ_NET_RATING'] = df.groupby(['TEAM_ID', 'Season', 'Basic_Type']).MINxNET.transform('sum')

df['MINxOFF'] = (df.MIN * df.OFF_RATING) / (g.MIN.transform("sum"))
df['ADJ_OFF_RATING'] = df.groupby(['TEAM_ID', 'Season', 'Basic_Type']).MINxOFF.transform('sum')

df['MINxDEF'] = (df.MIN * df.DEF_RATING) / (g.MIN.transform("sum"))
df['ADJ_DEF_RATING'] = df.groupby(['TEAM_ID', 'Season', 'Basic_Type']).MINxDEF.transform('sum')
```

```python
# Create a pivot table to tabulate means for each basic type
df_MEAN = pd.pivot_table(df, columns=['Basic_Type'], values=['ADJ_NET_RATING', 'ADJ_OFF_RATING', 'ADJ_DEF_RATING'], 
                         aggfunc=np.mean)
df_MEAN = df_MEAN.reindex(index = ['ADJ_OFF_RATING', 'ADJ_DEF_RATING', 'ADJ_NET_RATING']).round(3)

# Plot the figure
ax = df_MEAN.plot(kind='bar', figsize=(14, 10))

# Add annotations
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
df_MEAN.head()
```
