```python
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline
import seaborn as sns

from sklearn import preprocessing
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

from sklearn.mixture import GaussianMixture as GMM
from sklearn.cluster import KMeans, SpectralClustering as Spectral, AgglomerativeClustering as AC
from sklearn.metrics import silhouette_samples, silhouette_score

import renders as rs
import reduce_data as rd
```

```python
# Load Data
df = pd.DataFrame.from_csv('Data.csv')
print list(df.columns)
```

```python
# DataFrame for Relevant Features
# Stats previously modified to be 'percentage of total' or per minute
df2 = df[['3PAr', 'AVG_SEC_PER_TOUCH', 'DRIVES', 'CATCH_SHOOT_FG3A_PCT', 'PassFGAr',
          'PULL_UP_FGA', 'AVG_DRIB_PER_TOUCH', 'BASKET_TOUCH_PASSES_PCT', 'DIST_FEET', \
          'FRONT_CT_TOUCHES', 'POST_TOUCHES', 'ELBOW_TOUCHES', 'PAINT_TOUCHES', 'CATCH_SHOOT_FGA', \
          'CATCH_SHOOT_FG3A', 'PULL_UP_FG3A', 'DRIVE_PASSES_PCT', 'TIME_OF_POSS', \
          'CATCH_SHOOT_FGA_PCT', 'OREB_CHANCES', '0-3 Shooting %FGA', '3-10 Shooting %FGA', \
          '10-16 Shooting %FGA', '16 <3 Shooting %FGA', '3P Shooting %FGA', 'Corner %3PA Shooting', \
          '0-3', '3-10', '10-16', '16-3PT', 'Corner3']]
df2 = df2.replace(np.inf, np.nan)
df2 = df2.replace(-np.inf, np.nan)
df2 = df2.fillna(0)
```

```python
# Fix position list in preparation for Linear Discriminant Analysis
positions = []
for i in df['Pos']:
    if '-' in i:
        i = i.split('-')[0]
        positions.append(i)
    else:
        positions.append(i)
print set(positions)
```

```python
# It's close, but more intuitive are results acheived when merging SGs and SFs as well as PFs and Cs
df['Pos'] = positions
positions = []
for i in df['Pos']:
    if i == 'SG':
        positions.append('SF')
    elif i == 'PF':
        positions.append('C')
    else:
        positions.append(i)
df['Pos'] = positions
```

```python
# Perform LDA
lda = LDA(n_components=2)
lda_results = lda.fit(df2, df['Pos']).transform(df2)
 
# Generate LDA results plot
lda_display_results = rs.lda_results(df2, lda)

print lda_display_results['Explained Variance'].cumsum()

lda_results = preprocessing.normalize(lda_results, norm='l2')
min_max_scaler = preprocessing.MinMaxScaler()
lda_results = min_max_scaler.fit_transform(lda_results)

reduced_data = pd.DataFrame(lda_results)
```

```python
for i in range(10, 1, -1):
    preds = Spectral(n_clusters=i).fit_predict(reduced_data)
    score = silhouette_score(reduced_data, preds)
    print "Silhouette coefficient for {} clusters: {:.3f}.".format(i, score)
```

```python
preds = Spectral(n_clusters=3).fit_predict(reduced_data)
```

```python
# Apply offensive positions 
df2['Position'] = preds
df2['Name'] = df['Name']
df2['PLAYER_ID'] = df['PLAYER_ID']
df2['Pos'] = df['Pos']
df2['Season'] = df['Season']
df2.loc[df2['Name'] == 'Jamal Murray']['Position']
BH_num = df2.loc[df2['Name'] == 'LeBron James'].reset_index(drop=True).iloc[0]['Position']
W_num = df2.loc[df2['Name'] == 'Avery Bradley'].reset_index(drop=True).iloc[0]['Position']
B_num = df2.loc[df2['Name'] == 'Rudy Gobert'].reset_index(drop=True).iloc[0]['Position']
positions = []
for i in df2['Position']:
    if i == BH_num:
        positions.append('Ball-Handler')
    elif i == W_num:
        positions.append('Wing')
    elif i == B_num:
        positions.append('Big')
    else:
        pass
df2['Offensive Position'] = positions
df['Offensive Position'] = positions
```

```python
# Repeat the process for defense
df2 = df[['DREB_CHANCES', 'DEF_RIM_FGA', 'FGA_GT_15', 'FGA_LT_10', 'Height', 'Weight']]

lda = LDA(n_components=2)
lda_results = lda.fit(df2, df['Pos']).transform(df2)   #fit_transform provides only 3 dimension visualization below
 
# Generate PCA results plot
lda_display_results = rs.lda_results(df2, lda)

print lda_display_results['Explained Variance'].cumsum()

lda_results = preprocessing.normalize(lda_results, norm='l2')
min_max_scaler = preprocessing.MinMaxScaler()
lda_results = min_max_scaler.fit_transform(lda_results)
```

```python
for i in range(10, 1, -1):
    preds = Spectral(n_clusters=i).fit_predict(lda_results)
    score = silhouette_score(lda_results, preds)
    print "Silhouette coefficient for {} clusters: {:.3f}.".format(i, score)
    
preds = Spectral(n_clusters=3).fit_predict(lda_results)
```

```python
df['Position2'] = preds
G_num = df.loc[df['Name'] == 'Avery Bradley'].reset_index(drop=True).iloc[0]['Position2']
F_num = df.loc[df['Name'] == 'Andre Iguodala'].reset_index(drop=True).iloc[0]['Position2']
B_num = df.loc[df['Name'] == 'Rudy Gobert'].reset_index(drop=True).iloc[0]['Position2']
positions = []
for i in df['Position2']:
    if i == G_num:
        positions.append('Guard')
    elif i == F_num:
        positions.append('Forward')
    elif i == B_num:
        positions.append('Post')
df2['Defensive Position'] = positions
df['Defensive Position'] = positions
```
